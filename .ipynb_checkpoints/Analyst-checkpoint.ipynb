```python
ids = {}
prev = 0
#_lat, _lon = 42.34998902466133,-71.07742577791214
for k in dataset:
    for item in k:
        for the_trip in k[item]['info']:
            #print(the_trip.keys())
            if type(the_trip)==dict and 'vehicle' in the_trip.keys():
                #lat,lon = the_trip['vehicle']['vehicle_lat'],the_trip['vehicle']['vehicle_lon']
                #loc = abs(lat-_lat)+abs(lon-_lon)
                _id = the_trip['trip_id']
                #ETA
                pred = int(the_trip['pre_dt'])
                curr = int(the_trip['vehicle']['vehicle_timestamp'])
                #time = schedule - curr
                eta = int(the_trip['pre_away'])
                arrive_flag = 0
                if eta <60:
                    arrive_flag = curr 
                if _id not in ids:
                    ids[str(_id)] = [(pred,eta,curr,arrive_flag)]
                elif (pred,eta,curr,arrive_flag) not in ids[str(_id)]:
                    ids[str(_id)].append((pred,eta,curr,arrive_flag))
```

```python
ids
```

```python
#a: prediction time, b: eta in sec, c: time at request, d: actual arrive time
result = {}

for item in ids:
    flag = False
    con = 0
    for (a,b,c,d) in ids[item]:
        if d != 0:
            con = max(con,d)
            actual = con
            flag = True
    for (a,b,c,d)in ids[item]:
        if flag:
            if item not in result:
                result[item] = [(a,b,(b-(actual-c)))]
            #'eta '+str(b)+'- (acutal '+str(actual)+' - time at request '+str(c)+' ) =',
            else:
                result[item].append((a,b,(b-(actual-c))))
        else:
            result[item]='insufficient data'
```
